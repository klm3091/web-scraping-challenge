```python
from bs4 import BeautifulSoup
from splinter import Browser
import pandas as pd
import requests
from selenium import webdriver
import time
```

```python
executable_path = {'executable_path':'/usr/local/bin/chromedriver'}
#browser = webdriver.Chrome(executable_path)
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# NASA Mars news
url = 'https://mars.nasa.gov/news/'
browser.visit(url)
```

```python
html = browser.html

soup = BeautifulSoup(html, 'html.parser')

news_title = soup.find('div', class_='content_title').find('a').text
news_p = soup.find('div', class_='article_teaser_body').text

print(news_title)
print(news_p)
```

```python
#featured image
image_url_featured = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(image_url_featured)
```

```python
html_image = browser.html

soup = BeautifulSoup(html_image, 'html.parser')

featured_image_url  = soup.find('article')['style'].replace('background-image: url(','').replace(');', '')[1:-1]

main_url = 'https://www.jpl.nasa.gov'

featured_image_url = main_url + featured_image_url

featured_image_url
```

```python
# Scraping through a twitter page that reports the daily weather of Martian Sols (days)
mars_weather_url = 'https://twitter.com/marswxreport?lang=en'
browser.visit(mars_weather_url)
time.sleep(5)
html = browser.html
weather_soup = BeautifulSoup(html, "html.parser")
print(weather_soup)
```

```python
#Sorts from the parent tag
sol_weather_result = weather_soup.find_all('article')
print(sol_weather_result)
```

```python
#Loop through the tweets for the Sol info
weather_list = []
for data in sol_weather_result:
    tweets = data.find("div")
    #print(tweets)
    for tweet in tweets.find("div", class_="css-901oao r-jwli3a r-1qd0xha r-a023e6 r-16dba41 r-ad9z0x r-bcqeeo r-bnwqim r-qvutc0"):
        weather_list.append(tweet.parent.text)
weather_list[0]
mars_weather = weather_list[0]
mars_weather
```

```python
hemispheres_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(hemispheres_url)
```

```python
html_hemispheres = browser.html

soup = BeautifulSoup(html_hemispheres, 'html.parser')
items = soup.find_all('div', class_='item')

mars_images = []

#main url
hemispheres_main_url = 'https://astrogeology.usgs.gov'

#loop 
for i in items: 
    
    title = i.find('h3').text
    
    partial_img = i.find('a', class_='itemLink product-item')['href']
    
    browser.visit(hemispheres_main_url + partial_img)
     
    partial_img = browser.html
    
    soup = BeautifulSoup( partial_img, 'html.parser')
    
    #larger images
    img_url = hemispheres_main_url + soup.find('img', class_='wide-image')['src']
    
    mars_images.append({"title" : title, "img_url" : img_url})
    
mars_images
```
